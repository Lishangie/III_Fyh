```python
# Импорт NumPy
import numpy as np
```

## Создание массивов

```python
# Создание массива из списка
a = np.array([1, 2, 3])
a
```

```python
# Двумерный массив
b = np.array([[1.5, 2, 3], [4, 5, 6]])
b
```

```python
# Массив нулей
np.zeros((2, 2, 3))
```

```python
# Массив единиц
np.ones((2, 3))
```

```python
# Создание последовательности с шагом
np.arange(10, 30, 5)
```

## Индексирование и срезы

```python
# Получение элемента
a[1]
```

```python
# Срез массива
a[1:3]
```

```python
# Извлечение из двумерного массива
b[0:2, 0:3]
```

## Векторные операции

```python
# Поэлементное сложение
x = np.array([10, 20, 30])
y = np.array([1, 2, 3])
x + y
```

```python
# Поэлементное умножение
x * y
```

## Статистические функции

```python
# Среднее значение
np.mean(x)
```

```python
# Сумма элементов
np.sum(x)
```

```python
# Максимальное значение
np.max(x)
```
